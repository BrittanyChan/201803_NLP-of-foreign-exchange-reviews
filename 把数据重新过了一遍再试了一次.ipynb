```python
# --*-- encoding: utf-8 --*--
import pandas as pd

df = pd.read_csv(r'0506all-groups.csv',sep=None,encoding='ANSI')
df = df.drop_duplicates()
df.head()
```

```python
data_index = df[(df[r'专家意见'].isin([0]))]
data = df.drop(data_index.index)
data
```

```python
data_clean = data.dropna()
data = data_clean.reset_index()
data
```

```python
# 预处理数据

train_data = []
label_train = []
traindata = []
for i in range(0,round(0.6*len(data_clean))):
    train_data.append(data.loc[i,[r'专家意见']])
    traindata.append(data.loc[i,[r'专家意见']].values)
    label_train.append(data.loc[i,[r'情绪']].values)

test_data = []
label_test = []
testdata = []
for i in range(round(0.6*len(data_clean)),len(data_clean)):
    test_data.append(data.loc[i,[r'专家意见']])
    testdata.append(data.loc[i,[r'专家意见']].values)
    label_test.append(data.loc[i,[r'情绪']].values)

# 预览数据
print (train_data[0], '\n')
print (test_data[0])
```

```python
from pandas.core.frame import DataFrame
    
testdata = {'专家意见' : testdata, '情绪': label_test}
testdata=DataFrame(testdata) #将字典转换成为数据框
testdata
```

```python
import jieba

with open('train_data2.txt') as f:
    document = f.read()
    
    document_decode = document.encode('utf-8').decode('utf-8')
    document_cut = jieba.cut(document_decode)
    #print  ' '.join(jieba_cut)  //如果打印结果，则分词效果消失，后面的result无法显示
    result = ' '.join(document_cut)
    result = result
    with open('./train_data3.txt', 'w') as f2:
        f2.write(result)
f.close()
f2.close()

with open('test_data2.txt') as f:
    document = f.read()
    
    document_decode = document.encode('utf-8').decode('utf-8')
    document_cut = jieba.cut(document_decode)
    #print  ' '.join(jieba_cut)  //如果打印结果，则分词效果消失，后面的result无法显示
    result = ' '.join(document_cut)
    result = result
    with open('./test_data3.txt', 'w') as f2:
        f2.write(result)
f.close()
f2.close()
```

```python
import re

with open('./train_data3.txt') as f3:
    res1 = f3.read().encode('utf-8').decode('utf-8')
    res1 = re.sub("[A-Za-z0-9\[\`\~\!\@\#\$\^\&\*\(\)\=\|\{\}\'\:\;\'\,\，\。\[\]\.\<\>\/\?\~\！\@\#\\\&\*\%]", "", res1)

with open('./test_data3.txt') as f4:
    res2 = f4.read().encode('utf-8').decode('utf-8')
    res2 = re.sub("[A-Za-z0-9\[\`\~\!\@\#\$\^\&\*\(\)\=\|\{\}\'\:\;\'\,\，\。\[\]\.\<\>\/\?\~\！\@\#\\\&\*\%]", "", res2)
```

```python
res1 = res1.splitlines()
res2 = res2.splitlines()
```

```python
res1 = [x.replace('专家意见','') for x in res1]
res2 = [y.replace('专家意见','') for y in res2]
```

```python
#从文件导入停用词表
stpwrdpath = 'stop_words.txt'
stpwrd_dic = open(stpwrdpath, 'rb')
stpwrd_content = stpwrd_dic.read()
#将停用词表转换为list  
stpwrdlst = stpwrd_content.splitlines()
stpwrd_dic.close()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# 合并训练和测试集以便进行TFIDF向量化操作
corpus = res1+res2
vector = TfidfVectorizer(stop_words=stpwrdlst)
tfidf = vector.fit_transform(corpus)
# print ('TF-IDF处理结束.')
print (tfidf)
```

```python
len_train = len(train_data)
len_test = len(test_data)
len_total = len_train + len_test

# 恢复成训练集和测试集部分
train_x = tfidf[:len_train]
test_x = tfidf[len_train:len_total]
print ('TF-IDF处理结束.')
```

```python
import time
from sklearn import metrics 
from sklearn.metrics import precision_score, recall_score, accuracy_score
import numpy as np  
  
# Multinomial Naive Bayes Classifier  
def naive_bayes_classifier(train_x, label_train):  
    from sklearn.naive_bayes import MultinomialNB  
    model = MultinomialNB(alpha=0.01)  
    model.fit(train_x, label_train)  
    return model  
  
# KNN Classifier  
def knn_classifier(train_x, label_train):  
    from sklearn.neighbors import KNeighborsClassifier  
    model = KNeighborsClassifier()  
    model.fit(train_x, label_train)  
    return model  
  
# Logistic Regression Classifier  
def logistic_regression_classifier(train_x, label_train):  
    from sklearn.linear_model import LogisticRegression  
    model = LogisticRegression(penalty='l2')  
    model.fit(train_x, label_train)  
    return model  
  
# Random Forest Classifier  
def random_forest_classifier(train_x, label_train):  
    from sklearn.ensemble import RandomForestClassifier  
    model = RandomForestClassifier(n_estimators=8)  
    model.fit(train_x, label_train)  
    return model  
  
# Decision Tree Classifier  
def decision_tree_classifier(train_x, label_train):  
    from sklearn import tree  
    model = tree.DecisionTreeClassifier()  
    model.fit(train_x, label_train)  
    return model  
  
# GBDT(Gradient Boosting Decision Tree) Classifier  
def gradient_boosting_classifier(train_x, label_train):  
    from sklearn.ensemble import GradientBoostingClassifier  
    model = GradientBoostingClassifier(n_estimators=200)  
    model.fit(train_x, label_train)  
    return model  
  
# SVM Classifier  
def svm_classifier(train_x, label_train):  
    from sklearn.svm import SVC  
    model = SVC(kernel='rbf', probability=True)  
    model.fit(train_x, label_train)  
    return model  
  
# SVM Classifier using cross validation  
def svm_cross_validation(train_x, label_train):  
    from sklearn.grid_search import GridSearchCV  
    from sklearn.svm import SVC  
    model = SVC(kernel='rbf', probability=True)  
    param_grid = {'C': [1e-3, 1e-2, 1e-1, 1, 10, 100, 1000], 'gamma': [0.001, 0.0001]}  
    grid_search = GridSearchCV(model, param_grid, n_jobs = 1, verbose=1)  
    grid_search.fit(train_x, label_train)  
    best_parameters = grid_search.best_estimator_.get_params()  
    for para, val in best_parameters.items():  
        print (para, val)  
    model = SVC(kernel='rbf', C=best_parameters['C'], gamma=best_parameters['gamma'], probability=True)  
    model.fit(train_x, label_train)  
    return model  

if __name__ == '__main__':  
    
    test_classifiers = ['NB', 'KNN', 'LR', 'RF', 'DT', 'SVM', 'GBDT']  
    classifiers = {'NB':naive_bayes_classifier,'KNN':knn_classifier,'LR':logistic_regression_classifier,
                   'RF':random_forest_classifier,'DT':decision_tree_classifier,'SVM':svm_classifier,
                   'SVMCV':svm_cross_validation,'GBDT':gradient_boosting_classifier}

    for classifier in test_classifiers:
        print('******************* %s ********************' % classifier)
        start_time = time.time()
        model = classifiers[classifier](train_x, label_train)
        print('training took %fs!' % (time.time() - start_time))
        predict = model.predict(test_x)
        precision = metrics.precision_score(label_test, predict)
        recall = metrics.recall_score(label_test, predict)
        print('precision: %.2f%%, recall: %.2f%%' % (100 * precision, 100 * recall))
        accuracy = metrics.accuracy_score(label_test, predict)
        print('accuracy: %.2f%%' % (100 * accuracy))
```
