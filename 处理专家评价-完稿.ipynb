```python
# 第一步 ：将文本数据导入，进行简单清洗
```

```python
# --*-- encoding: utf-8 --*--
import pandas as pd

df = pd.read_csv(r'0328all-groups.csv',sep=None,encoding='ANSI')
df = df.drop_duplicates()
df.head()
```

```python
data_index = df[(df[r'专家意见'].isin([0]))]
data = df.drop(data_index.index)
data
```

```python
data_clean = data.dropna()
data = data_clean.reset_index()
data
```

```python
# 第二步：准备将文本数据导入txt文档中进行分词处理同时把百分之六十分成training set另外百分之二十设为test set
```

```python
# 预处理数据

train_data = []
label_train = []
traindata = []
for i in range(0,round(0.6*len(data_clean))):
    train_data.append(data.loc[i,[r'专家意见']])
    traindata.append(data.loc[i,[r'专家意见']].values)
    label_train.append(data.loc[i,[r'情绪']].values)

test_data = []
label_test = []
testdata = []
for i in range(round(0.6*len(data_clean)),len(data_clean)):
    test_data.append(data.loc[i,[r'专家意见']])
    testdata.append(data.loc[i,[r'专家意见']].values)
    label_test.append(data.loc[i,[r'情绪']].values)

# 预览数据
print (train_data[0], '\n')
print (test_data[0])
```

```python
# 第三步：生成新的testdata的dataframe，为后面测试时做准备
```

```python
from pandas.core.frame import DataFrame
    
testdata = {'专家意见' : testdata, '情绪': label_test}
testdata=DataFrame(testdata) #将字典转换成为数据框
testdata
```

```python
# 第五步：分词结巴处理
```

```python
import jieba

with open('train_data.txt') as f:
    document = f.read()
    
    document_decode = document.encode('utf-8').decode('utf-8')
    document_cut = jieba.cut(document_decode)
    #print  ' '.join(jieba_cut)  //如果打印结果，则分词效果消失，后面的result无法显示
    result = ' '.join(document_cut)
    result = result
    with open('./train_data2.txt', 'w') as f2:
        f2.write(result)
f.close()
f2.close()

with open('test_data.txt') as f:
    document = f.read()
    
    document_decode = document.encode('utf-8').decode('utf-8')
    document_cut = jieba.cut(document_decode)
    #print  ' '.join(jieba_cut)  //如果打印结果，则分词效果消失，后面的result无法显示
    result = ' '.join(document_cut)
    result = result
    with open('./test_data2.txt', 'w') as f2:
        f2.write(result)
f.close()
f2.close()
```

```python
# 第六步：将分好词的文档导入进来，做正则匹配保留有用的文本信息
```

```python
import re

with open('./train_data2.txt') as f3:
    res1 = f3.read().encode('utf-8').decode('utf-8')
    res1 = re.sub("[A-Za-z0-9\[\`\~\!\@\#\$\^\&\*\(\)\=\|\{\}\'\:\;\'\,\，\。\[\]\.\<\>\/\?\~\！\@\#\\\&\*\%]", "", res1)

with open('./test_data2.txt') as f4:
    res2 = f4.read().encode('utf-8').decode('utf-8')
    res2 = re.sub("[A-Za-z0-9\[\`\~\!\@\#\$\^\&\*\(\)\=\|\{\}\'\:\;\'\,\，\。\[\]\.\<\>\/\?\~\！\@\#\\\&\*\%]", "", res2)
```

```python
res1 = res1.splitlines()
res2 = res2.splitlines()
```

```python
res1 = [x.replace('专家意见','') for x in res1]
res2 = [y.replace('专家意见','') for y in res2]
```

```python
# 第七步：导入停用词，带入TF-IDF模型
```

```python
#从文件导入停用词表
stpwrdpath = 'stop_words.txt'
stpwrd_dic = open(stpwrdpath, 'rb')
stpwrd_content = stpwrd_dic.read()
#将停用词表转换为list  
stpwrdlst = stpwrd_content.splitlines()
stpwrd_dic.close()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# 合并训练和测试集以便进行TFIDF向量化操作
corpus = res1+res2
vector = TfidfVectorizer(stop_words=stpwrdlst)
tfidf = vector.fit_transform(corpus)
# print ('TF-IDF处理结束.')
print (tfidf)
```
