```python
# --*-- encoding: utf-8 --*--
import pandas as pd
data_train_offline = pd.read_csv(r'data/ccf_offline_stage1_train.csv')
data_test_offline = pd.read_csv(r'data/ccf_offline_stage1_test_revised.csv')
```

```python
# 第一步：导入数据
```

```python
res2 = pd.merge(data_train_offline, data_test_offline, on=['User_id', 'Merchant_id'])
res2 = res2.drop(['Coupon_id_x', 'Discount_rate_x', 'Distance_x', 'Date_received_x', 'Coupon_id_y',
               'Discount_rate_y', 'Date_received_y'], axis=1)
res2['Distance_y'] = res2['Distance_y'].replace('null', 'unknown')
res2
```

```python
# 第二步：准备清洗数据，把两张表按照User_id和Merchant_id都一样的拼在一起，并且取出指定的列
```

```python
res3 = res2[res2['Date'] != 'null']
res3 = res3.drop_duplicates('Date',keep='first') 
res3.head(10)
```

```python
# 第三步：去掉里面没有购买的记录，再去掉重复的行（该表得出的是某线下用户购买某商品的具体日期以及距离）
```

```python
res4 = pd.merge(res3, data_test_offline, on=['User_id', 'Merchant_id'])
res_final = res4.drop(['Date', 'Merchant_id', 'Distance_y', 'Discount_rate', 'Distance'], axis=1)
res_final['Probability'] = 1
res_final
```

```python
# 第四步：这里根据特定用户购买了特定商品为合并表的条件，除去了result里面不需要的列，并且加上了可能性（全为 1）
# 我这里认为一个人是否购买主要取决于指定的用户是否会买指定的商品，与其他的无关
```

```python
all_res = data_test_offline[['User_id', 'Coupon_id', 'Date_received']]
all_res
```

```python
# 第五步：选取出最终提交result的三列
```

```python
summary = pd.merge(all_res, res_final, on=['User_id', 'Coupon_id', 'Date_received'], how='left')
summary = summary.fillna(0)
summary
```

```python
# 第六步：合并两个表（以outer的形式）
```

```python
summary.to_csv('result.csv',index=False,sep=',')
```
