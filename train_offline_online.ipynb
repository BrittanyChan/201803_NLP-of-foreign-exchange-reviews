```python
# --*-- encoding: utf-8 --*--
import pandas as pd
data_train_online = pd.read_csv(r'ccf_online_stage1_train.csv')
data_train_offline = pd.read_csv(r'ccf_offline_stage1_train.csv')
```

```python
res = pd.merge(data_train_online, data_train_offline, on=['User_id'])
res = res.drop(['Action', 'Coupon_id_x', 'Discount_rate_x', 'Date_received_x', 'Coupon_id_y',
               'Discount_rate_y', 'Date_received_y'], axis=1)
res.head()
```

```python
res['Date_x_new'] = res['Date_x'].replace('null', 0)
res['Date_y_new'] = res['Date_y'].replace('null', 0)
res = res.drop(['Date_x', 'Date_y'], axis=1)
res
```
